```python
!pip install fuzzywuzzy
from fuzzywuzzy import fuzz, process
import pandas as pd

# Load the first Excel file into a pandas dataframe
df1 = pd.read_excel('/content/chenotsi.xlsx')

# Load the second Excel file into a pandas dataframe
df2 = pd.read_excel('/content/chenotsi2.xlsx')

# Extract the columns containing the lists you want to compare
list1 = df1['Stok Name'].tolist()
list2 = df2['Stok Name'].tolist()

# Add another column to extract from both dataframes
column_to_print = 'Stock Code'
column1 = df1['Stok No'].tolist()
column2 = df2['Stok No'].tolist()

# Find the common elements between the two lists using fuzzy string matching
common_elements = []
for i, elem1 in enumerate(list1):
    matches = process.extract(elem1, list2, scorer=fuzz.token_sort_ratio, limit=10)
    for elem2, score in matches:
        if score >= 60:  # adjust this threshold as needed
            common_elements.append((elem1, elem2, column1[i], column2[list2.index(elem2)], score))

# Print the common elements with scores and the additional column
for elem in common_elements:
    print(f"{elem[0]} and {elem[1]} have a similarity score of {elem[4]} and {column_to_print} of {elem[2]} and {elem[3]}")

```
